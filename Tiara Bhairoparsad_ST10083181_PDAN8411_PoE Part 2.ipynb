# What is classification?
According to Wolff, R. 2020., classification in machine learning is a method where a model learns from a given dataset to predict the category of new data points. This process is a form of supervised learning, meaning the model is trained on a dataset with known labels (training data) and then evaluated on a separate dataset (test data) before being used to classify new, unseen data. The model uses this training data to understand the relationship between the input features and the output labels, allowing it to classify new, unseen data into the appropriate categories. Classification involves recognizing, understanding, and grouping ideas or objects into predefined categories or sub-populations. Classification is a type of "pattern recognition where algorithms analyze training data to identify patterns (such as similar words, sentiments, or number sequences) and use these patterns to categorize future data sets.

Classification algorithms use labeled training data to learn the patterns and relationships within the data. This training enables the model to predict the likelihood that new data will belong to one of the predetermined categories. For example, one common application is email filtering, where emails are classified as "spam" or "non-spam” which involves training a model on labeled email data to identify patterns and characteristics that distinguish spam from legitimate emails. Classification falls under the broader category of supervised learning. Classification algorithms are part of supervised learning methods, meaning that predictions are made based on labeled historical data. In supervised learning, the model is trained using a dataset where each instance is associated with a predefined label or category. This contrasts with unsupervised learning, where there are no predefined labels or categories, and the goal is to identify patterns or groupings within the data based solely on its inherent structure. In supervised learning, categories are established based on common attributes, and the model learns to predict the labels for new data based on these learned patterns. Additionally, within supervised learning, there is a distinction between classification and regression algorithms. Classification predicts categorical outcomes, while regression aims to predict continuous, real-valued outcomes. Unsupervised learning, on the other hand, involves tasks like clustering and association, where the objective is to explore the data and find hidden structures without any prior knowledge of the categories. Anon. n.d., states that classification can be performed on both structured data (like tables) and unstructured data (like text or images). The primary objective is to identify the class or category to which new data belongs, based on similarities to the training data.

Classification algorithms are the methods by which software learns to autonomously categorize data into different groups. These algorithms analyze the input data to identify patterns and common characteristics, enabling the prediction of the data's category.
Some commonly used classification algorithms according to Anon. 2023., includes:
- Decision Trees: Use a tree-like model of decisions and their consequences. They are easy to understand and interpret.
- Random Forests: An ensemble method that uses multiple decision trees to improve classification accuracy.
- Support Vector Machines (SVM): Find the hyperplane that best separates the classes in the feature space.
- Naive Bayes: Based on Bayes' theorem, it assumes independence between predictors.
- K-Nearest Neighbors (KNN): Classifies a data point based on how its neighbors are classified.

There are 3 types of classification:
- Binary Classification: Involves two classes, e.g., "spam" and "non-spam."
- Multiclass Classification: Involves more than two classes, such as categorizing emails into "work," "personal," "spam," and "promotions."
- Multilabel Classification: Each instance can belong to multiple categories simultaneously, such as tagging a news article with multiple topics like "politics," "economy," and "health."

The key Concepts in Classification include:
- Label: The category or class assigned to a data point. For example, in an email classification task, labels could be "spam" or "not spam."
- Features: The attributes or characteristics of the data used to make predictions. In the context of email classification, features might include the presence of certain keywords, the sender’s email address, or the length of the email.
- Training Data: A subset of the dataset used to train the model, containing input features and corresponding labels.
- Test Data: A subset of the dataset used to evaluate the model’s performance, which also includes input features and labels but is not used during training.

According to Agrawal, K. 2024., There are many ways for measuring classification performance. Performance Metrics are as follows:
- Accuracy: The ratio of correctly predicted instances to the total instances.
- Precision and Recall: Precision is the ratio of correctly predicted positive observations to the total predicted positives, while recall is the ratio of correctly predicted positive observations to all observations in the actual class.
- F1 Score: The harmonic mean of precision and recall, useful when there is an uneven class distribution.
- Confusion Matrix: A table used to evaluate the performance of a classification model, showing the true vs. predicted classifications.

While classification is powerful, it comes with several challenges according to Anon. 2022:
- Imbalanced Data: When one class is significantly more frequent than others, it can bias the model's predictions.
- Overfitting: When the model performs well on training data but poorly on test data due to capturing noise rather than the underlying pattern.
- Feature Selection: Identifying the most relevant features from potentially high-dimensional data can be complex.
- Scalability: Managing large datasets efficiently requires optimized algorithms and sufficient computational resources.

## How would changing the training/test splits will affect your predictions and accuracy?
The training/test split ratio refers to the proportion of the dataset allocated to training the model versus testing its performance. Changing the training/test splits can affect predictions and accuracy by:
1.	Higher Training Data Ratio:
- Effect on Predictions: Increasing the proportion of data allocated to training (e.g., 80% or more) allows the model to learn from more examples, potentially improving its ability to capture underlying patterns and relationships in the data.
- Effect on Accuracy: With more training data, the model may generalize better to unseen examples, leading to potentially higher accuracy on the testing set. However, there's a risk of overfitting if the model becomes too complex and starts memorizing the training data instead of learning generalizable patterns.
2.	Higher Testing Data Ratio:
- Effect on Predictions: Increasing the proportion of data allocated to testing (e.g., 40% or more) provides more instances for evaluating the model's performance on unseen data. This can help provide a more robust assessment of the model's generalization capabilities.
- Effect on Accuracy: With more testing data, the accuracy estimate may become more reliable as it is based on a larger sample size. However, if the training data is insufficient, the model may not learn enough to make accurate predictions, leading to lower accuracy.
3.	Balanced Training/Test Split:
- Effect on Predictions: A balanced split ratio (e.g., 50/50) ensures an equal distribution of data for training and testing. This can help prevent biases in the model's performance evaluation and ensure a fair assessment of its predictive capabilities.
- Effect on Accuracy: The accuracy estimate may be more representative of the model's true performance when the training and testing sets are balanced. However, the model may not have enough data to learn complex patterns if the split ratio is too balanced, potentially leading to lower accuracy.

```python
# 2. Comparing the Gaussian Naive Bayes Classifier to K-NN Algorithm

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'C:/Users/lab_services_student/Desktop/Naive-Bayes-Classification-Data.csv'
data = pd.read_csv(file_path)

# Check for missing values
print(data.isnull().sum())

# Feature and target variables
X = data[['glucose', 'bloodpressure']]
y = data['diabetes']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Train the Gaussian Naïve Bayes Classifier
gnb = GaussianNB()
gnb.fit(X_train, y_train)

# Predictions
y_pred_gnb = gnb.predict(X_test)
y_prob_gnb = gnb.predict_proba(X_test)[:, 1]  # Predicted probabilities for the positive class

# Evaluate the model
print("Gaussian Naïve Bayes Classifier Performance:")
print("Accuracy:", accuracy_score(y_test, y_pred_gnb))
print("Classification Report:\n", classification_report(y_test, y_pred_gnb))
print("ROC AUC Score:", roc_auc_score(y_test, y_prob_gnb))

# Hyperparameter tuning for K-Nearest Neighbors (K-NN)
param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}

grid_search = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

# Best parameters and best score
best_params = grid_search.best_params_
print(f"Best Parameters for K-NN: {best_params}")

# Train the K-NN classifier with the best parameters
best_knn = grid_search.best_estimator_
best_knn.fit(X_train, y_train)

# Predictions
y_pred_knn = best_knn.predict(X_test)
y_prob_knn = best_knn.predict_proba(X_test)[:, 1]  # Predicted probabilities for the positive class

# Evaluate the model
print("K-Nearest Neighbors (K-NN) Algorithm Performance:")
print("Accuracy:", accuracy_score(y_test, y_pred_knn))
print("Classification Report:\n", classification_report(y_test, y_pred_knn))
print("ROC AUC Score:", roc_auc_score(y_test, y_prob_knn))

# Cross-validation for both models
cv_scores_gnb = cross_val_score(gnb, X, y, cv=10, scoring='accuracy')
cv_scores_knn = cross_val_score(best_knn, X, y, cv=10, scoring='accuracy')

print(f"Gaussian Naïve Bayes Cross-Validation Accuracy: {np.mean(cv_scores_gnb)}")
print(f"K-NN Cross-Validation Accuracy: {np.mean(cv_scores_knn)}")

#3 Plot ROC Curves
fpr_gnb, tpr_gnb, _ = roc_curve(y_test, y_prob_gnb)
fpr_knn, tpr_knn, _ = roc_curve(y_test, y_prob_knn)

plt.figure(figsize=(10, 6))
plt.plot(fpr_gnb, tpr_gnb, label='Gaussian Naïve Bayes')
plt.plot(fpr_knn, tpr_knn, label='K-NN')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) curve')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), shadow=True, ncol=2)
plt.show()

# Summary of comparison
print("\nSummary of Model Comparison:")
print(f"Naïve Bayes Classifier Accuracy: {accuracy_score(y_test, y_pred_gnb)}")
print(f"K-NN Classifier Accuracy: {accuracy_score(y_test, y_pred_knn)}")
print(f"Naïve Bayes ROC AUC Score: {roc_auc_score(y_test, y_prob_gnb)}")
print(f"K-NN ROC AUC Score: {roc_auc_score(y_test, y_prob_knn)}")
print(f"Naïve Bayes Cross-Validation Accuracy: {np.mean(cv_scores_gnb)}")
print(f"K-NN Cross-Validation Accuracy: {np.mean(cv_scores_knn)}")

# Code Attribution:
# Anon. n.d. sklearn.metrics.plot_roc_curve. scikit-learn. [Online] Available at: https://scikit-learn.org/0.24/modules/generated/sklearn.metrics.plot_roc_curve.html [Accessed 17 May 2024]
```

# 3. Receiver Operating Characteristic (ROC) curve:
- The ROC curve is a plot of the true positive rate (sensitivity) against the false positive rate (1 - specificity) for different threshold values.
It illustrates the trade-off between sensitivity and specificity as the classification threshold is varied. 
- Labels for the x-axis (False Positive Rate) and y-axis (True Positive Rate) are set using plt.xlabel() and plt.ylabel() respectively.
- The x-axis and y-axis limits are set to [0.0, 1.0] for both axes
- A diagonal dashed line ([0, 1], [0, 1]) is plotted representing the ROC curve of a random classifier.

- The ROC curve's steep upward slope near the top-left corner indicates that the classifier achieves high sensitivity (True Positive Rate) while maintaining a low false positive rate. This implies that the NB classifier is highly effective at identifying diabetic patients while minimizing misclassifications of non-diabetic patients.
- The AUC score for K-NN is 0.978, which is slightly lower than that of NB but still very high. This suggests that the K-NN algorithm also has excellent discriminatory power, albeit slightly lower than NB.

- The true positive rate (TPR) is plotted on the y-axis. It represents the proportion of actual positive cases that are correctly classified as positive by the model. TPR is calculated as TP / (TP + FN), where TP is the number of true positives and FN is the number of false negatives.
- The false positive rate (FPR) is plotted on the x-axis. It represents the proportion of actual negative cases that are incorrectly classified as positive by the model. FPR is calculated as FP / (FP + TN), where FP is the number of false positives and TN is the number of true negatives.
- AUC quantifies the overall performance of the classifier. It represents the area under the ROC curve, with values ranging from 0 to 1. A perfect classifier has an AUC of 1, while a random classifier has an AUC of 0.5. Higher AUC values indicate better discriminative ability of the model, where it separates positive and negative cases more effectively.

# 5. Precision-Recall Curve:
- The x-axis is labeled as 'Recall', representing the true positive rate.
- The y-axis is labeled as 'Precision', representing the positive predictive value
- The Precision-Recall curve is a plot of precision (positive predictive value) against recall (true positive rate) for different threshold values. It illustrates the trade-off between precision and recall as the classification threshold is varied.
- Precision is the ratio of true positive predictions to the total number of positive predictions. In the context of binary classification, it represents the ability of the classifier to correctly identify positive samples (e.g., diabetic patients) without misclassifying negative samples (e.g., non-diabetic patients) as positive. A higher precision indicates fewer false positives.
- Recall (Sensitivity): Recall is the ratio of true positive predictions to the total number of actual positive samples. It measures the classifier's ability to correctly identify all positive samples. A higher recall indicates fewer false negatives.
- Precision-Recall Curve: This curve illustrates the trade-off between precision and recall for different decision thresholds. Each point on the curve corresponds to a different threshold used to classify samples as positive or negative. Generally, as the threshold decreases, recall increases while precision decreases, and vice versa.
- In this graph:
  - Gaussian Naïve Bayes (GNB): The Precision-Recall curve for GNB (blue line) shows how precision and recall change for different decision thresholds. At lower recall values, precision is relatively high, indicating that GNB can effectively identify true positives with minimal false positives. As recall increases, precision decreases gradually, suggesting a trade-off between precision and recall. Overall, GNB achieves a balance between precision and recall across various thresholds.
  - K-Nearest Neighbors (KNN): The Precision-Recall curve for KNN (green line) follows a similar pattern to GNB, showing the trade-off between precision and recall. KNN tends to have slightly higher precision at lower recall values compared to GNB, indicating it may be more conservative in labeling positive samples. However, as recall increases, precision decreases, suggesting a similar trade-off between precision and recall as observed with GNB.

```python
#9 Learning curves to compare the performance of the Gaussian Naïve Bayes Classifier and the K-Nearest Neighbors 
from sklearn.model_selection import learning_curve

# Define a function to plot learning curves
def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None, n_jobs=None, train_sizes=np.linspace(.1, 1.0, 5)):
    plt.figure(figsize=(10, 6))
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes, scoring='accuracy')
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    # Move the legend to under the x-axis
    plt.legend(loc="upper center", bbox_to_anchor=(0.5, -0.15), shadow=True, ncol=2)

    return plt

# Plot learning curves for Gaussian Naïve Bayes Classifier
title = "Learning Curves (Gaussian Naïve Bayes)"
plot_learning_curve(gnb, title, X_train, y_train, ylim=(0.7, 1.01), cv=5, n_jobs=-1)

# Plot learning curves for K-Nearest Neighbors (K-NN) Algorithm
title = "Learning Curves (K-NN)"
plot_learning_curve(best_knn, title, X_train, y_train, ylim=(0.7, 1.01), cv=5, n_jobs=-1)

plt.show()

# Code Attribution:
# Anon. n.d. Plotting Learning Curves and Checking Models’ Scalability. scikit-learn. [Online] Available at: https://scikit-learn.org/stable/auto_examples/model_selection/plot_learning_curve.html [Accessed 21 May 2024]


```

```python
#10 Decision boundary visualization for both models
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier

# Load the dataset
file_path = 'C:/Users/lab_services_student/Desktop/Naive-Bayes-Classification-Data.csv'
data = pd.read_csv(file_path)

# Feature and target variables
X = data[['glucose', 'bloodpressure']]
y = data['diabetes']

# Create instances of classifiers
gnb = GaussianNB()
knn = KNeighborsClassifier(n_neighbors=5)

# Fit classifiers to data
gnb.fit(X, y)
knn.fit(X, y)

# Define meshgrid range
x_min, x_max = X['glucose'].min() - 1, X['glucose'].max() + 1
y_min, y_max = X['bloodpressure'].min() - 1, X['bloodpressure'].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

# Create meshgrid DataFrame with feature names
meshgrid_df = pd.DataFrame(np.c_[xx.ravel(), yy.ravel()], columns=['glucose', 'bloodpressure'])

# Predict classes for meshgrid points
Z_gnb = gnb.predict(meshgrid_df)
Z_gnb = Z_gnb.reshape(xx.shape)

Z_knn = knn.predict(meshgrid_df)
Z_knn = Z_knn.reshape(xx.shape)

# Create subplot grid
fig, axarr = plt.subplots(1, 2, figsize=(12, 5))

# Plot decision boundaries for GNB and KNN
for idx, Z, title in zip([0, 1], [Z_gnb, Z_knn], ['Decision Boundary for Gaussian Naïve Bayes', 'Decision Boundary for K-Nearest Neighbors']):
    # Plot decision boundary
    axarr[idx].contourf(xx, yy, Z, alpha=0.4, cmap=ListedColormap(['#FFAAAA', '#AAAAFF']))
    
    # Plot data points for 'Non-Diabetic' class
    axarr[idx].scatter(X.loc[y == 0, 'glucose'], X.loc[y == 0, 'bloodpressure'], c='r', label='Non-Diabetic', edgecolor='k', s=20)
    
    # Plot data points for 'Diabetic' class
    axarr[idx].scatter(X.loc[y == 1, 'glucose'], X.loc[y == 1, 'bloodpressure'], c='b', label='Diabetic', edgecolor='k', s=20)
    
    # Set plot titles and axis labels
    axarr[idx].set_title(title)
    axarr[idx].set_xlabel('Glucose')
    axarr[idx].set_ylabel('Blood Pressure')
    
    # Add legend below the x-axis
    axarr[idx].legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), shadow=True, ncol=2)

plt.show()

# Code Attribution:
# Deepthi, R. 2019. KNN visualization in just 13 lines of code. scikit-learn. [Online] Available at: https://towardsdatascience.com/knn-visualization-in-just-13-lines-of-code-32820d72c6b6 [Accessed 23 May 2024]
```
